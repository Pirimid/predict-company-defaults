## Z-Score

```python
cols_of_interest = ["Total Current Assets", "Total Liabilities", "Net Income After Taxes", "Total Equity",
                    "Accounts Receivable - Trade, Gross","Total Current Liabilities", "Revenue", "Working Capital", 
                   "Net Income Before Taxes"]
```

```python
def compute_z_score(data):
    x1 = data['Working Capital'] / data['Total Current Assets']
    x2 = data['Net Income After Taxes'] / data['Total Current Assets']
    x3 = data['Net Income Before Taxes'] / data['Total Current Assets']
    x4 = data['Total Equity'] / data['Total Liabilities']
    x5 = data['Accounts Receivable - Trade, Gross'] / data['Total Current Assets']
    
    return (001.2 * x1) + (001.4 * x2) + (003.3 * x3) + (000.6 * x4) + (1.0 * x5)
    
```

```python
# COMPANY_NAMES = ["Coral India Finance and Housing Ltd", "Sintex Industries Ltd"]
```

```python
revenue = None

revenue_cols = []
for col in data.columns:
    for int_col in cols_of_interest:
        if col.find(int_col) >= 0:
            revenue_cols.append(col)
            
company_data = {}
grouped = data.groupby('Company Name')
for name, group in tqdm(grouped):
    average_data = {}
    for int_col in cols_of_interest:
        data_int = []
        for rev_col in revenue_cols:
            if rev_col.find(int_col) >= 0:
                col_data = group[rev_col].astype(str).str.replace(",", "").replace(' -   ', str(0)).str.replace("%", "").astype(float).values
                data_int.append(col_data[0].mean())
        average_data[int_col] = np.mean(data_int[0:15])
    company_data[name] = compute_z_score(average_data)
```

```python
# company_data
```

```python
z_score = []
for company in data["Company Name"]:
    for key,value in company_data.items():
        if key == company:
            z_score.append(value)

print(data['Company Name'][-10:])
z_score[-10:]
```

```python
data['Z Score'] = z_score
data.head()
```

```python
data['Discrimination'] = np.where(data['Z Score'] > 2.99, 'Safe Zone', 'Distress Zone')
# data['Discrimination'].head()
```
